```python
# Assimetria -  mede o grau de simetria da curva, em relação a distribuição normal.
df.skew()
```

Para dados normalmente distribuídos, a assimetria é bem próxima de zero.

Uma assimetria positiva, com valor > 0, significa que a cauda da distribuição está mais para a direita. Ou seja, que os valores de moda < mediana < média.

Uma assimetria negativa, com valor < 0, significa que a cauda da distribuição está mais para a esquerda. Ou seja, que os valores de média < mediana < moda.

```python
# Curtose - mede o grau de achatamento da curva, em relação a curva normal
df.kurtosis()
```

Se o valor encontrado é igual 0, você tem uma distribuição mesocúrtica; se maior que 0, leptocúrtica, correspondendo a uma curva mais pontuda; se menor que 0, platicúrtica, ou seja, uma curva mais achatada, com maior variabilidade dos dados.

# O coeficiente de correlação

```python
correlation = df.corr()
```

```python
# Matriz de correlação
import seaborn as sn
plot = sn.heatmap(correlation, annot = True, fmt=".1f", linewidths=.6)
plt.savefig("Matriz de correlação.jpg")
plot
```

Observe no gráfico que os pontos(fatores) dentro da sombra estão muito próximos a linha ou sobre ela, isso significa que estes pontos(fatores) são os mais significativos, ou seja, sua correlação é positiva e muito forte. 

Assim os pontos(fatores) ques estão fora da sombra, alguns estão próximos a ela e um está um pouco distante, isso significa que estes pontos(fatores) são menos signiticativos, mas que ainda apresentam uma correlação positiva, porém fraca.
