```python
# Extraindo modelo de dados utilizados pelo CENIPA
import urllib.request
import sys

try:
  urllib.request.urlretrieve("http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/modelo_dados.png", "modelo_dados.png")
  print("Imagem salva! =)")
except:
  erro = sys.exc_info()
  print("Ocorreu um erro:", erro)
```

```python
# Extraindo os dados selecionados do CENIPA que serão manipulados no projeto "Tabela de ocorrência" 

import urllib.request
import sys

try:
  urllib.request.urlretrieve("http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/ocorrencia.csv", "ocorrencia.csv")
  print("Tabela salva! =)")
except:
  erro = sys.exc_info()
  print("Ocorreu um erro:", erro)
```

```python
import pandas as pd
```

```python
# Visualizando os dados da tabela de ocorrência do CENIPA
data = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
data.head(10)
```

```python
# Salvando dataframe extraído para a validação dos dados 
data.to_csv('ocorrencia_data_extraido.csv')
data.head()
```
