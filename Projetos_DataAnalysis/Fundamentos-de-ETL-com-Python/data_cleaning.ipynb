```python
# Iniciando a limpeza definitiva para a transformação dos dados
data = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
data = data.drop(['codigo_ocorrencia1', 'codigo_ocorrencia3','codigo_ocorrencia4', 'ocorrencia_latitude', 'ocorrencia_longitude', 
          'ocorrencia_pais', 'investigacao_aeronave_liberada','investigacao_status', 'divulgacao_relatorio_numero',
          'divulgacao_relatorio_publicado', 'divulgacao_dia_publicacao','total_aeronaves_envolvidas', 
          'ocorrencia_saida_pista'], axis=1)
data.head()
```

```python
# limpando(Removendo) linhas sem dados informados
data.dropna()
```
