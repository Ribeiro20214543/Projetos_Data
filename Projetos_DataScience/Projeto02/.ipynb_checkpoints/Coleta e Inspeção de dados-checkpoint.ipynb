## Importando os dados

```python
import chardet
import pandas as pd
```

```python
# arquivo
cars_train = 'cars_train.csv'
cars_test = 'cars_test.csv'
```

```python
# verifica a codificacao dos arquivos

with open(cars_train, 'rb') as f:
    result1 = chardet.detect(f.read())
    
with open(cars_test, 'rb') as f:
    result2 = chardet.detect(f.read())
    
print("O encoding do arquivo ", cars_train,"é:", result1['encoding'])
print("O encoding do arquivo ", cars_test,"é:", result2['encoding'])
```

```python
# abre os arquivos

df_test = pd.read_csv('cars_test.csv', sep = '\t', encoding='utf-16')
df_train = pd.read_csv('cars_train.csv', sep = '\t', encoding='utf-16')
```
